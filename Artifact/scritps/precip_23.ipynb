```python
def regrid_snodas(wrf_file, snodas_file, var, multiple_time = True):
    
    if multiple_time:
        ds_snodas= snodas_file.swap_dims({'time': 'XTIME'})
    else:   
        ds_snodas = snodas_file.sel(time = '2022-01-12T06:00:00')
        
    ds_snodas = snodas_file[var]

    grid_template = xr.open_dataset(wrf_file)
    LAT = grid_template.variables['XLAT'][0].copy()
    LON =grid_template.variables['XLONG'][0].copy()

    #create target grid
    target_grid = xr.Dataset({'lat': LAT,'lon': LON})

    # Create xESMF regridder using the target grid
    regridder = xe.Regridder(ds_snodas, target_grid, 'bilinear')
    return regridder(ds_snodas)
```

```python
def prep_taylor(MOR_ERA,MOR_CFS,WD_ERA,WD_CFS,WS_ERA,WS_CFS, TH_ERA,TH_CFS,sn_regrid_sh,type):
    
    if type == 'SNOWH':
        th_cfs_spatial_mean = TH_CFS['SNOWH'].mean(dim = ('south_north', 'west_east'))*1e3
        th_era_spatial_mean = TH_ERA['SNOWH'].mean(dim = ('south_north', 'west_east'))*1e3
        mor_cfs_spatial_mean = MOR_CFS['SNOWH'].mean(dim = ('south_north', 'west_east'))*1e3
        mor_era_spatial_mean = MOR_ERA['SNOWH'].mean(dim = ('south_north', 'west_east'))*1e3
        wsm6_cfs_spatial_mean = WS_CFS['SNOWH'].mean(dim = ('south_north', 'west_east'))*1e3
        wsm6_era_spatial_mean = WS_ERA['SNOWH'].mean(dim = ('south_north', 'west_east'))*1e3
        wdm6_cfs_spatial_mean = WD_CFS['SNOWH'].mean(dim = ('south_north', 'west_east'))*1e3
        wdm6_era_spatial_mean = WD_ERA['SNOWH'].mean(dim = ('south_north', 'west_east'))*1e3
        snodas_sp_mean = sn_regrid_sh.mean(dim = ('south_north', 'west_east'))
        
    else:
        th_cfs_spatial_mean = TH_CFS['SNOW'].mean(dim = ('south_north', 'west_east'))
        th_era_spatial_mean = TH_ERA['SNOW'].mean(dim = ('south_north', 'west_east'))
        mor_cfs_spatial_mean = MOR_CFS['SNOW'].mean(dim = ('south_north', 'west_east'))
        mor_era_spatial_mean = MOR_ERA['SNOW'].mean(dim = ('south_north', 'west_east'))
        wsm6_cfs_spatial_mean = WS_CFS['SNOW'].mean(dim = ('south_north', 'west_east'))
        wsm6_era_spatial_mean = WS_ERA['SNOW'].mean(dim = ('south_north', 'west_east'))
        wdm6_cfs_spatial_mean = WD_CFS['SNOW'].mean(dim = ('south_north', 'west_east'))
        wdm6_era_spatial_mean = WD_ERA['SNOW'].mean(dim = ('south_north', 'west_east'))
        snodas_sp_mean = sn_regrid_sh.mean(dim = ('south_north', 'west_east'))


    samples = [[float(np.std(th_cfs_spatial_mean)),np.corrcoef(snodas_sp_mean,th_cfs_spatial_mean)[0,1], "TH-CFS"],  
            [float(np.std(th_era_spatial_mean)),np.corrcoef(snodas_sp_mean,th_era_spatial_mean)[0,1], "TH-ERA"],  
            [float(np.std(mor_cfs_spatial_mean)),np.corrcoef(snodas_sp_mean,mor_cfs_spatial_mean)[0,1], "MOR-CFS"], 
            [float(np.std(mor_era_spatial_mean)),np.corrcoef(snodas_sp_mean, mor_era_spatial_mean)[0,1], "MOR-ERA"],
            [float(np.std(wsm6_cfs_spatial_mean)),np.corrcoef(snodas_sp_mean, wsm6_cfs_spatial_mean)[0,1], "WS-CFS"],
            [float(np.std(wsm6_era_spatial_mean)),np.corrcoef(snodas_sp_mean, wsm6_era_spatial_mean)[0,1], "WS-ERA"],
            [float(np.std(wdm6_era_spatial_mean)),np.corrcoef(snodas_sp_mean, wdm6_era_spatial_mean)[0,1], "WD-ERA"],
            [float(np.std(wdm6_cfs_spatial_mean)),np.corrcoef(snodas_sp_mean, wdm6_cfs_spatial_mean)[0,1], "WD-CFS"]
              ]


    return float(np.std(snodas_sp_mean)), samples
```

```python
def plot_grouped_histogram(df,title,ax,legend = True, xlabel = True):
    count_dict = {}

    for col in df.columns[1:]:
        # Initializing counters for each category
        count_cat1 = 0
        count_cat2 = 0
        count_cat3 = 0
        count_cat4 = 0
        for value in df[col]:
            if value >= 0.75:
                count_cat1 += 1
            elif 0.65 <= value < 0.75:
                count_cat2 += 1
            elif 0.5 <= value < 6.5:
                count_cat3 += 1
            else:
                count_cat4 += 1
                
        count_dict[col] = {'Category 1': count_cat1, 'Category 2': count_cat2, 'Category 3': count_cat3, 'Category 4': count_cat4}

    # Plotting the grouped histogram
    data = count_dict
    categories = ['Category 1', 'Category 2', 'Category 3', 'Category 4']
    values = {category: [data[key][category] for key in data.keys()] for category in categories}
    keys = list(data.keys())

    # Set the width of the bars
    barWidth = 0.15

    # Set position of bar on X axis
    r1 = np.arange(len(values['Category 1']))
    r2 = [x + barWidth for x in r1]
    r3 = [x + barWidth for x in r2]
    r4 = [x + barWidth for x in r3]

    #plt.figure(figsize=(8, 3))

    # Create the grouped bar plot
    ax.bar(r1, values['Category 1'], color='g', width=barWidth, edgecolor='grey', label=r'Very Good')
    ax.bar(r2, values['Category 2'], color='y', width=barWidth, edgecolor='grey', label=r'Good')
    ax.bar(r3, values['Category 3'], color='b', width=barWidth, edgecolor='grey', label=r'Satisfactory')
    ax.bar(r4, values['Category 4'], color='r', width=barWidth, edgecolor='grey', label=r'Unsatisfatory')

    # Adding labels and ticks
    
    
    if xlabel: 
        ax.set_xlabel('Experiment', fontweight='bold')
        ax.set_xticks([r + 1.5 * barWidth for r in range(len(values['Category 1']))], keys, rotation=90)


    else:
        ax.set_xticks([])
        ax.set_xticklabels([])
    
    ax.set_title(title,fontweight='bold')
    ax.set_ylabel('Count',fontweight='bold')
    # Placing the legend outside the plot
    if legend:
        ax.legend(loc='upper left', bbox_to_anchor=(1, 1))
```

```python


# Create a 1x3 figure
fig, ax = plt.subplots(2, 1, figsize=(14, 8))


plot_grouped_histogram(swh,'Snow Water Equivalent',ax[0],xlabel = False)
plot_grouped_histogram(swe,'Snow Height',ax[1],legend = False,xlabel = True)
plt.show()
```
