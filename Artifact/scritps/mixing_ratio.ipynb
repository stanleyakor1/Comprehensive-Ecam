```python
from packages import *
from domain import *
```

```python
path = '/bsuhome/stanleyakor/scratch/2022/'

thom = xr.open_dataset(path + 'Thompson_radiation_22_daily.nc')
mor = xr.open_dataset(path + 'Morrison_radiation_22_daily.nc')
wsm6 = xr.open_dataset(path + 'WSM6_radiation_22_daily.nc')
wdm6 = xr.open_dataset(path + 'WDM6_radiation_22_daily.nc')

start = '2021-10-01'
end = '2022-09-30'
```

```python
thom  = thom.sel(XTIME=slice(start, end))
mor = mor.sel(XTIME=slice(start, end))
wsm6 = wsm6.sel(XTIME=slice(start, end))
wdm6 = wdm6.sel(XTIME=slice(start, end))
```

## HYDROMETEOR MIXING RATIO

```python
#rain
th_rain = thom['QRAIN'].mean(dim = ('XTIME','south_north','west_east'))
mor_rain = mor['QRAIN'].mean(dim = ('XTIME','south_north','west_east'))
wsm6_rain = wsm6['QRAIN'].mean(dim = ('XTIME','south_north','west_east'))
wdm6_rain = wdm6['QRAIN'].mean(dim = ('XTIME','south_north','west_east'))
```

```python
#snow
th_snow = thom['QSNOW'].mean(dim = ('XTIME','south_north','west_east'))
mor_snow = mor['QSNOW'].mean(dim = ('XTIME','south_north','west_east'))
wsm6_snow = wsm6['QSNOW'].mean(dim = ('XTIME','south_north','west_east'))
wdm6_snow = wdm6['QSNOW'].mean(dim = ('XTIME','south_north','west_east'))
```

```python
#ice

th_ice = thom['QICE'].mean(dim = ('XTIME','south_north','west_east'))
mor_ice = mor['QICE'].mean(dim = ('XTIME','south_north','west_east'))
wsm6_ice = wsm6['QICE'].mean(dim = ('XTIME','south_north','west_east'))
wdm6_ice = wdm6['QICE'].mean(dim = ('XTIME','south_north','west_east'))
```

```python
#graupel

th_graup = thom['QGRAUP'].mean(dim = ('XTIME','south_north','west_east'))
mor_graup = mor['QGRAUP'].mean(dim = ('XTIME','south_north','west_east'))
wsm6_graup = wsm6['QGRAUP'].mean(dim = ('XTIME','south_north','west_east'))
wdm6_graup = wdm6['QGRAUP'].mean(dim = ('XTIME','south_north','west_east'))
```

```python
#vapor

th_vapor = thom['QVAPOR'].mean(dim = ('XTIME','south_north','west_east'))
mor_vapor = mor['QVAPOR'].mean(dim = ('XTIME','south_north','west_east'))
wsm6_vapor = wsm6['QVAPOR'].mean(dim = ('XTIME','south_north','west_east'))
wdm6_vapor = wdm6['QVAPOR'].mean(dim = ('XTIME','south_north','west_east'))
```

```python
#cloud

th_cloud = thom['QCLOUD'].mean(dim = ('XTIME','south_north','west_east'))
mor_cloud = mor['QCLOUD'].mean(dim = ('XTIME','south_north','west_east'))
wsm6_cloud = wsm6['QCLOUD'].mean(dim = ('XTIME','south_north','west_east'))
wdm6_cloud = wdm6['QCLOUD'].mean(dim = ('XTIME','south_north','west_east'))
```

```python

x_values = np.arange(1, 45, 1) #model level

plt.figure(figsize=(9, 8))

# Create empty lists to collect handles and labels for the combined legend
legend_handles = []
legend_labels = []

# Subplot 1
plt.subplot(3, 3, 1)
plt.plot(th_rain * 1e3, x_values, 'r', label='Thompson QRAIN')
plt.plot(mor_rain * 1e3, x_values, 'k', label='Morrison QRAIN')
plt.plot(wsm6_rain * 1e3, x_values, 'g', label='WSM6 QRAIN')
plt.plot(wdm6_rain * 1e3, x_values, 'm', label='WDM6 QRAIN')
plt.xlim(0, 0.0025)
plt.gca().xaxis.set_major_formatter(plt.FormatStrFormatter('%.3f'))
plt.ylabel('Y Label')
plt.title('QRAIN')
legend_handles.append(plt.gca().get_legend_handles_labels()[0])
legend_labels.append('Thompson')
plt.ylabel('Model level') 

# Subplot 2
plt.subplot(3, 3, 2)
plt.plot(th_snow * 1e3, x_values, 'r', label='Thompson QSNOW')
plt.plot(mor_snow * 1e3, x_values, 'k', label='Morrison QSNOW')
plt.plot(wsm6_snow * 1e3, x_values, 'g', label='WSM6 QSNOW')
plt.plot(wdm6_snow * 1e3, x_values, 'm', label='WDM6 QSNOW')
plt.xlim(0, 0.02)
plt.title('QSNOW')
plt.gca().xaxis.set_major_formatter(plt.FormatStrFormatter('%.3f'))
legend_handles.append(plt.gca().get_legend_handles_labels()[0])
legend_labels.append('Morrison')
plt.tick_params(axis='y', labelleft=False)  # Hide y-axis ticks and labels

# Subplot 3
plt.subplot(3, 3, 3)
plt.plot(th_ice * 1e3, x_values, 'r', label='Thompson QICE')
plt.plot(mor_ice * 1e3, x_values, 'k', label='Morrison QICE')
plt.plot(wsm6_ice * 1e3, x_values, 'g', label='WSM6 QICE')
plt.plot(wdm6_ice * 1e3, x_values, 'm', label='WDM6 QICE')
plt.xlim(0, 0.007)
plt.gca().xaxis.set_major_formatter(plt.FormatStrFormatter('%.3f'))
plt.xlabel('X Label')
plt.title('QICE')
legend_handles.append(plt.gca().get_legend_handles_labels()[0])
legend_labels.append('WSM6')
plt.tick_params(axis='y', labelleft=False) 

# Subplot 4
plt.subplot(3, 3, 4)
plt.plot(th_graup * 1e3, x_values, 'r', label='Thompson QGRAUP')
plt.plot(mor_graup * 1e3, x_values, 'k', label='Morrison QGRAUP')
plt.plot(wsm6_graup * 1e3, x_values, 'g', label='WSM6 QGRAUP')
plt.plot(wdm6_graup * 1e3, x_values, 'm', label='WDM6 QGRAUP')
plt.xlim(0, 0.0015)
plt.gca().xaxis.set_major_formatter(plt.FormatStrFormatter('%.3f'))
plt.xlabel('X Label')
plt.title('QGRAUP')
plt.ylabel('Model level')
legend_handles.append(plt.gca().get_legend_handles_labels()[0])
legend_labels.append('WDM6')
plt.xlabel('Mixing ratio (g/kg)')

# Subplot 5
plt.subplot(3, 3, 5)
plt.plot(th_vapor * 1e3, x_values, 'r', label='Thompson QVAPOR')
plt.plot(mor_vapor * 1e3, x_values, 'k', label='Morrison QVAPOR')
plt.plot(wsm6_vapor * 1e3, x_values, 'g', label='WSM6 QVAPOR')
plt.plot(wdm6_vapor * 1e3, x_values, 'm', label='WDM6 QVAPOR')

plt.gca().xaxis.set_major_formatter(plt.FormatStrFormatter('%.3f'))
plt.title('QVAPOR')
legend_handles.append(plt.gca().get_legend_handles_labels()[0])
legend_labels.append('WSM6')
plt.xlabel('Mixing ratio (g/kg)')
plt.tick_params(axis='y', labelleft=False) 
# Subplot 6
plt.subplot(3, 3, 6)
plt.plot(th_cloud * 1e3, x_values, 'r', label='Thompson QCLOUD')
plt.plot(mor_cloud * 1e3, x_values, 'k', label='Morrison QCLOUD')
plt.plot(wsm6_cloud * 1e3, x_values, 'g', label='WSM6 QCLOUD')
plt.plot(wdm6_cloud * 1e3, x_values, 'm', label='WDM6 QCLOUD')
plt.gca().xaxis.set_major_formatter(plt.FormatStrFormatter('%.3f'))
plt.xlabel('X Label')
plt.title('QCLOUD')
legend_handles.append(plt.gca().get_legend_handles_labels()[0])
legend_labels.append('WDM6')
plt.xlabel('Mixing ratio (g/kg)')
plt.tick_params(axis='y', labelleft=False) 

# Create a single legend outside the last figure
plt.figlegend(
    handles=legend_handles[-1],
    labels=legend_labels,
    loc='upper center',
    bbox_to_anchor=(0.5,1.05),
    ncol=len(legend_labels)
)

plt.tight_layout()
plt.show()
```
