# Compare KGE score across the SNOTEL sites for all experiments

<hr style="border-width:4px; border-color:coral">

The Kling-Gupta Efficiency (KGE) is a metric used to evaluate the performance of hydrological models, particularly in the context of streamflow simulation. It was proposed by Kling et al. in 2012 as a modification of the Nash-Sutcliffe Efficiency (NSE) to address some of its limitations. The Kling-Gupta Efficiency is defined as:

$KGE = 1 - \sqrt{ (r - 1)^2 + (\beta - 1)^2 + (\gamma - 1)^2 }$

where:

r is the correlation coefficient between observed and simulated values.

 is the ratio of the standard deviation of simulated values to the standard deviation of observed values.

 is the ratio of the mean of simulated values to the mean of observed values.
The KGE ranges from 
 to 1, with a higher value indicating better model performance. A KGE of 1 indicates perfect model performance, and a value of 0 indicates that the model's performance is no better than a model that predicts the mean of the observed values.

The KGE incorporates three components: correlation, variability, and bias. This makes it a more comprehensive metric than the Nash-Sutcliffe Efficiency, which is solely based on the correlation between observed and simulated values.


In this assessment, we classify KGE into four categories:

$\textbf{1}$ "Very Good" for KGE values greater than or equal to 0.75

$\textbf{3}$ "Good" for KGE values between 0.65 and 0.75

$\textbf{3}$ "Satisfactory" for KGE values between 0.5 and 0.65

$\textbf{4}$ "Unsatisfactory" for KGE values less than 0.5
